```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

data = pd.read_csv('C:\\Users\\nabhe\\OneDrive\\Desktop\\appds\\user_behavior_dataset.csv')
stats = data.describe()
print("Stats : ", stats[['Data Usage (MB/day)', 'Age', 'Battery Drain (mAh/day)']])

#medians:
print("medians: ", data[['Data Usage (MB/day)', 'Age', 'Battery Drain (mAh/day)']].median())

#skewness
print("skewness: ", data[['Data Usage (MB/day)', 'Age', 'Battery Drain (mAh/day)']].skew())

#kurtosis
print("kurtosis: ", data[['Data Usage (MB/day)', 'Age', 'Battery Drain (mAh/day)']].kurt())

#correlation between 'Data Usage (MB/day)' & 'Age'
correlation = data['Data Usage (MB/day)'].corr(data['Age'])
print("Correlation between Data Usage & Age is ", correlation)

correlation = data['Data Usage (MB/day)'].corr(data['Battery Drain (mAh/day)'])
print("Correlation between Data Usage and Battery Drain is ", correlation)

```

```python
import matplotlib.pyplot as plt
import pandas as pd 
data = pd.read_csv('C:\\Users\\nabhe\\OneDrive\\Desktop\\appds\\user_behavior_dataset.csv')

#correlation between Data usage and Battery drain
x = data['Battery Drain (mAh/day)']
y = data['Data Usage (MB/day)']

def plot_scatter(x, y):
    """
    creates scatter plot of x(Battery Drain) and y(Data usage) 
    """
    plt.figure(dpi= 144)

    #scatter plot
    plt.scatter(x, y, color = 'red')

    plt.xlabel('Battery Drain (mAh/day)')
    plt.ylabel('Data Usage (MB/day)')
    plt.title('Data Usage vs Battery Drain')
    
    plt.show()
    return

plot_scatter(x, y)
```

```python
import matplotlib.pyplot as plt
import pandas as pd 

data = pd.read_csv('C:\\Users\\nabhe\\OneDrive\\Desktop\\appds\\user_behavior_dataset.csv')
r = data.select_dtypes(include= 'number').corr()

#heatmap plot showing correlation between different variables
def plot_heatmap(r):
    """
    creates heatmap plot using correlation values of different variables
    """
    plt.figure(dpi= 144)
    plt.imshow(r, cmap = 'viridis', interpolation = 'nearest')
    plt.xticks(range(8), r.index, rotation= 90, fontdict= dict(fontsize= 8))
    plt.yticks(range(8), r.index, fontdict= dict(fontsize= 8))

    #putting correlation values in each box of heatmap plot
    for i in range(8):
        for j in range(8):
            plt.text(i, j, "{:.2f}".format(r.values[i,j]), ha= 'center')
            
    plt.colorbar()
    plt.title('Correlation Heatmap', fontdict= dict(fontweight= 'bold'))

    plt.show()
    return
    
plot_heatmap(r)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:\\Users\\nabhe\\OneDrive\\Desktop\\appds\\user_behavior_dataset.csv')

#calculating mean of data usage according to device model and gender
new_data = data.groupby(['Device Model', 'Gender'])['Data Usage (MB/day)'].mean().unstack()

def plot_bar():
    """
    creates bar plot using average data usage according to gender on different device models
    """
    plt.figure(dpi= 144)
    new_data.plot(kind='bar', width= 0.7)
    plt.ylabel('Average Data Usage (MB/day)')
    plt.xlabel('Device Models')
    plt.title('Average Data Usage (MB/day) by Device Model and Gender')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title= 'Gender', loc= 'best')
    plt.tight_layout()
    
    plt.show()
    return

plot_bar()
```
